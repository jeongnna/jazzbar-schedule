```python
import numpy as np
import pandas as pd
```

```python
schedule = pd.read_csv("schedule.csv")
```

```python
with open("test.txt") as f:
    content = f.readlines()
members = [x.strip().split() for x in content]
```

```python
def print_info(schedule, members):
    date, stage, team = schedule['date'], schedule['stage'], schedule['team']
    members = ", ".join(members)
    info = 'Date   : %s\nStage  : %s\nTeam   : %s\nMembers: %s' % (date, stage, team, members)
    return info

def when_my_star_performs(name, schedule, members):
    out = [print_info(schedule.iloc[i], m) for i, m in enumerate(members) if name in m]
    out = "\n\n".join(out)
    return out
```

```python
out = when_my_star_performs('심규민', schedule, content)
print(out)
```
